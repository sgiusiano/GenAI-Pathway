```python
import pdf_optimized_rag_chatbot as prc
```

```python
chat = prc.PDFChatbot()
default_url = "https://app.readthedocs.org/projects/requests/downloads/pdf/latest/"
chat.load_pdf(default_url, collection_name="requests_docs")
```

```python
docs_retriever_similarity = chat.retrieve_context(
    query="How does request handle authentication?",
    search_type="similarity",
    re_rank_compress=False,
    use_metadata=False
    )

docs_retriever_similarity
```

```python
docs_retriever_mmr = chat.retrieve_context(
    query="How does request handle authentication?",
    search_type="mmr",
    re_rank_compress=False,
    use_metadata=False
    )

docs_retriever_mmr
```

```python
docs_retriever_mmr_compress = chat.retrieve_context(
    query="How does request handle authentication?",
    search_type="mmr",
    re_rank_compress=True,
    use_metadata=False
    )

docs_retriever_mmr_compress
```

```python
docs_retriever_mmr_compress_filter_metadata = chat.retrieve_context(
    query="How does request handle authentication?",
    search_type="mmr",
    re_rank_compress=True,
    use_metadata=True
    )

docs_retriever_mmr_compress_filter_metadata
```

```python
retrievers = {
    "similarity": docs_retriever_similarity,
    "mmr": docs_retriever_mmr,
    "mmr_compress": docs_retriever_mmr_compress,
    "mmr_compress_filter_metadata": docs_retriever_mmr_compress_filter_metadata
}
```

```python
for doc in docs_retriever_similarity:
    #print(doc.page_content)
    print(doc.id)
```

```python
for key, docs in retrievers.items():
    print(f"\n--- Retriever: {key} ---")
    for doc in enumerate(docs):
        doc_id = doc.id if hasattr(doc, 'id') else doc[1].id
        doc_page_content = doc.page_content if hasattr(doc, 'page_content') else doc[1].page_content
        doc_metadata = doc.metadata if hasattr(doc, 'metadata') else doc[1].metadata
        print(f"Document ID: {doc_id}")
        print(f"Content: {doc_page_content[:10]}...")  # Print first
        print(f"Metadata: {doc_metadata}")
```
